# What time of day are the accidents most frequent in?

```python
NH_df = filtered_df[filtered_df['City'] == 'New Haven'].copy()

# Convert 'Start_Time' to datetime format
NH_df['Start_Time'] = pd.to_datetime(NH_df['Start_Time'])

# Extract the hour from 'Start_Time' and format it in 12-hour format with AM/PM
NH_df.loc[:, 'Hour'] = NH_df['Start_Time'].dt.strftime('%I:%M %p')

# Count the number of accidents for each hour
accidents_by_hour = NH_df['Hour'].value_counts().sort_index()

# Plot the results
plt.figure(figsize=(12, 6))
accidents_by_hour.plot(kind='bar', color='skyblue')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Accidents')
plt.title('Accidents by Hour of the Day in New Haven (12-hour format)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

NH_df = filtered_df[filtered_df['City'] == 'New Haven'].copy()
# Convert 'Start_Time' to datetime format
NH_df['Start_Time'] = pd.to_datetime(NH_df['Start_Time'])

# Extract the day of the week from 'Start_Time'
NH_df['Day_of_Week'] = NH_df['Start_Time'].dt.dayofweek

# Map numeric day of week to actual day names
day_names = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
NH_df['Day_of_Week'] = NH_df['Day_of_Week'].map(lambda x: day_names[x])

# Count the number of accidents for each day of the week
accidents_by_day = NH_df['Day_of_Week'].value_counts().reindex(day_names)

# Plot the results
plt.figure(figsize=(10, 6))
accidents_by_day.plot(kind='bar', color='skyblue')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Accidents')
plt.title('Accidents by Day of the Week')
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

# Assuming 'df' is your DataFrame
# Replace 'your_dataset.csv' with the actual path or name of your dataset
# df = pd.read_csv('your_dataset.csv')

# Select a reduced set of features and target variable
features = ['Distance(mi)','Pressure(in)','Humidity(%)','Wind_Chill(F)','Temperature(F)','Wind_Speed(mph)','Amenity', 'Bump', 'Crossing','Junction','Railway',
                'Stop','Traffic_Signal','Sunrise_Sunset']

target = 'Severity'

# Subset the DataFrame with selected features and target variable
df_subset = filtered_df[features + [target]]

# Handle missing values
df_subset = df_subset.dropna()

# Convert boolean columns to integers if applicable
bool_columns = ['Amenity', 'Bump', 'Crossing', 'Give_Way', 'Junction', 'No_Exit', 'Railway',
                'Roundabout', 'Station', 'Stop', 'Traffic_Calming', 'Traffic_Signal', 'Turning_Loop']
filtered_df[bool_columns] = filtered_df[bool_columns].astype(int)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df_subset[features], df_subset[target], test_size=0.2, random_state=42)

# Choose a model
model = RandomForestClassifier()

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred))
accuracy = accuracy_score(y_test, y_pred)

# Print the accuracy score
print("Accuracy:", accuracy)
```
